```python
import pandas as pd
import pandas
```

```python
# 깃허브의 www.github.com/hyowakn -> python_lec
csFilePath = "https://raw.githubusercontent.com/hyokwan/python-lecture/refs/heads/master/dataset/kopo_customerdata.csv"
customerData = pandas.read_csv(csFilePath)
customerData
```

```python
# customerData에서 TOTAL_AMOUNT를 리스트로 뽑아내서 customerTotalLost라는 변수에 넣음
customerTotalLost = list( customerData.TOTAL_AMOUNT )
customerTotalLost

# customerTotalLost에서 min과 max로 최소값과 최대값을 추출 하고 각각 min_customer_data와 max_customer_data라는 변수에 넣음
min_customer_data= min(customerTotalLost)
max_customer_data= max(customerTotalLost)

# min_customer_data와 max_customer_data를 리스트화 하여 각각 min_except와 max_except라는 변수에 넣어줌
min_except = [min_customer_data]
max_except = [max_customer_data]

# customerTotalLost에서 min_except의 값과 max_except값을 제거해줌
customerTotalLost.remove(min_except[0])
customerTotalLost.remove(max_except[0])

# customerTotalLost의 합과 리스트 내부 요소의 길이(갯수)를 구해 sum_customer_data와 len_customer_data라는 변수에 넣어줌
sum_customer_data = sum(customerTotalLost)
len_customer_data = len(customerTotalLost)

# sum_customer_data와 len_customer_data로 평균을 내고 int함수를 이용해 정수로 바꾼뒤 result변수에 넣어줌
average = sum_customer_data / len_customer_data
result = int(average)

result
```
